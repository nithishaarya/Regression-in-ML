## Support Vector Machine Using Gradient Descent


```python
import numpy as np
import matplotlib.pyplot as plt
```

Dataset
 X   Y
-2   4
 4   1
 1   6
 2   4
 6   2

```python
#storing data in single dataset
X = np.array([[-2, 4, -1], [4, 1, -1], [1, 6, -1], [2, 4, -1], [6, 2, -1]]) 
y = np.array([-1, -1, 1, 1, 1])
```

```python
for d, i in enumerate(X):
    if d<2:
        plt.scatter(i[0], i[1], s=120, marker='_', linewidths=2)
    else:
        plt.scatter(i[0], i[1], s=120, marker='+', linewidths=2)
#scatter plot of the data
```

```python
def svm_gd_plot(X, Y):
    w = np.zeros(len(X[0])) #Weight of the column for three columns
    learning_rate = 1
    epochs = 100000
    errors = [] #to track misclassification

    #gradient descent
    for epoch in range(1, epochs):
        error = 0 #error status 
        for i, j in enumerate(X):
            #misclassification condition
            if(Y[i]*np.dot(X[i], w)) < 1:
                #weight update - misclassification
                w = w + learning_rate * ((X[i] * Y[i]) + (-2 * (1/epoch)*w))
                error = 1 
            else:
                #proper classification & its weight udpate
                w = w + learning_rate * (-2 *(1/epoch) * w )
            errors.append(error)
    return w
```

```python
for d, i in enumerate(X):
    #plotting of - ve data
    if d<2:
        plt.scatter(i[0], i[1], s=120, marker='_', linewidths=2)
    else:
        plt.scatter(i[0], i[1], s=120, marker='+', linewidths=2)

plt.scatter(2, 2, s= 120, marker='_', linewidths=2, color='brown') # Testing Data Plots
plt.scatter(4,3, s=120, marker='+', linewidths=2, color='blue')

w = svm_gd_plot(X, y)

#printing the hyper plane
x2 = [w[0], w[1], -w[1], w[0]]
x3 = [w[0], w[1], w[1], -w[0]]

x2x3 = np.array([x2,x3])
X, Y, U, V = zip(*x2x3)
plt.quiver(X, Y, U, V, scale=1, color='blue')
plt.show()
```
